```python
medals=winter.groupby(['Athlete','Medal'])['Sport'].count().reset_index().sort_values(by='Sport',ascending=False)
medals=medals.drop_duplicates(subset=['Medal'],keep='first')
medals.columns=[['Athlete','Medal','Count']]
medals
```

```python
winter.Athlete.unique()
```

```python
medal=pd.merge(winter,countries,left_on='Country',right_on='Code',how='left')
medal=medal[['Year','City','Sport','Discipline','Athlete','Country_x','Gender','Event','Medal','Country_y']]
medal.columns=['Year','City','Sport','Discipline','Athlete','Code','Gender','Event','Medal','Country']
medal
```

```python
medals_map = medal.groupby(['Country','Code'])['Medal'].count().reset_index().sort_values(by='Medal',ascending=False)
medals_map.head()
```

```python
ax=medal.groupby(['Country'])['Medal'].count().reset_index().sort_values(by='Metal',ascending=False)[:11]
ax
```

```python
medals_country=medal.groupby(['Country','Medal'])['Code'].count().reset_index().sort_values(by='Code',ascending=False)
medals_country=medals_country.pivot('Country','Medal','Code').fillna(0)
top=medals_country.sort_values(by='Gold',ascending=False)[:11]
top.plot.barh(width=0.8,color=['#CD7F32','#FFDF00','#D3D3D3'])
```

```python
fig,ax=plt.subplots(1,2,figsize=(18,12))
men=medal[medal['Gender']=='Men']
men=men.groupby(['Athlete','Medal'])['Code'].count().reset_index().sort_values(by='Medal',ascending=False)
men=men[men['Athlete'].isin(winter['Athlete'].value_counts().index[:15])]
men=men.pivot('Athlete','Medal','Code').fillna(0)

men.plot(kind='barh',width=0.8,color=['#CD7F32','#FFDF00','#D3D3D3'],ax=ax[0])
ax[0].set_title('Best Male Athletes')
ax[0].set_ylabel('Athlete')


fmen=medal[medal['Gender']=='Women']
fmen=fmen.groupby(['Athlete','Medal'])['Code'].count().reset_index().sort_values(by='Medal',ascending=False)
fmen=fmen[fmen['Athlete'].isin(winter['Athlete'].value_counts().index[:15])]
fmen=fmen.pivot('Athlete','Medal','Code').fillna(0)

fmen.plot(kind='barh',width=0.8,color=['#CD7F32','#FFDF00','#D3D3D3'],ax=ax[1])
ax[1].set_title('Best FEMale Athletes')
ax[1].set_ylabel('Athlete')
```

```python
medal_men=medal[(medal['Code']=='USA') & (medal['Gender']=='Men')]
medal_men=medal_men.groupby(['Year','Medal'])['Event'].count().reset_index()
fig,ax=plt.subplots(2,figsize=(15,8))
medal_men=medal_men.pivot('Year','Medal','Event').fillna(0)
medal_men.plot(color=['#CD7F32','#FFDF00','#D3D3D3'],ax=ax[0])
ax[0].set_xlabel(' ')
ax[0].set_title('Performance Of USA Men')

medal_fmen=medal[(medal['Code']=='USA') & (medal['Gender']=='Women')]
medal_fmen=medal_fmen.groupby(['Year','Medal'])['Event'].count().reset_index()
medal_fmen=medal_fmen.pivot('Year','Medal','Event').fillna(0)
medal_fmen.plot(color=['#CD7F32','#FFDF00','#D3D3D3'],ax=ax[1])
ax[1].set_xlabel(' ')
ax[1].set_title('Performance Of USA Men')
```
